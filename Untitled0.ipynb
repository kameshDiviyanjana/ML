```python
import os
import numpy as np
import librosa
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, LSTM, RepeatVector, TimeDistributed, Dense
import tensorflowjs as tfjs
```

```python
# Step 1: Load and Preprocess Audio Files
# def load_audio_files(directory, sr=16000, n_mfcc=13):
#     audio_features = []
#     for file in os.listdir(directory):
#         if file.endswith(".wav"):
#             file_path = os.path.join(directory, file)
#             audio, _ = librosa.load(file_path, sr=sr)
#             mfccs = librosa.feature.mfcc(audio, sr=sr, n_mfcc=n_mfcc).T  # Transpose to (timesteps, features)
#             audio_features.append(mfccs)
#     return audio_features
def load_audio_files(directory, sr=16000, n_mfcc=13):
    audio_features = []
    for file in os.listdir(directory):
        if file.endswith(".wav"):
            file_path = os.path.join(directory, file)
            audio, _ = librosa.load(file_path, sr=sr)
            # Extract MFCC features
            mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=n_mfcc).T  # Transpose to (timesteps, features)
            audio_features.append(mfccs)
    return audio_features


# Normalize audio sequences to the same length
def pad_sequences(sequences, max_len):
    padded_sequences = []
    for seq in sequences:
        if len(seq) < max_len:
            pad_width = max_len - len(seq)
            padded_seq = np.pad(seq, ((0, pad_width), (0, 0)), mode='constant')
        else:
            padded_seq = seq[:max_len]
        padded_sequences.append(padded_seq)
    return np.array(padded_sequences)
```

```python

# Step 2: Preprocess the Data
directory = "/content/drive/MyDrive/RPDATA/BESD"
audio_data = load_audio_files(directory)
max_timesteps = 100  # Set max sequence length
features = 13        # Number of MFCC features

# Pad and normalize sequences
audio_data_padded = pad_sequences(audio_data, max_timesteps)
scaler = MinMaxScaler()
audio_data_normalized = np.array([scaler.fit_transform(seq) for seq in audio_data_padded])
```

```python
# Step 3: Build the LSTM Autoencoder
def build_autoencoder(timesteps, features, latent_dim=32):
    # Encoder
    inputs = Input(shape=(timesteps, features))
    encoded = LSTM(64, return_sequences=False)(inputs)
    encoded = Dense(latent_dim, activation='relu')(encoded)

    # Decoder
    decoded = RepeatVector(timesteps)(encoded)
    decoded = LSTM(64, return_sequences=True)(decoded)
    outputs = TimeDistributed(Dense(features))(decoded)

    # Autoencoder
    autoencoder = Model(inputs, outputs)
    autoencoder.compile(optimizer='adam', loss='mse')
    return autoencoder
```

```python
# Build and train the autoencoder
latent_dim = 32
autoencoder = build_autoencoder(max_timesteps, features, latent_dim)
autoencoder.summary()
```

```python
# Train the autoencoder
autoencoder.fit(
    audio_data_normalized, audio_data_normalized,
    epochs=40, batch_size=32, validation_split=0.2
)
```

```python

# Step 4: Extract Latent Features
encoder = Model(autoencoder.input, autoencoder.layers[2].output)  # Extract encoder model
latent_features = encoder.predict(audio_data_normalized)
```

```python

# Step 5: Cluster the Latent Features
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
clusters = kmeans.fit_predict(latent_features)
```

```python

# Step 6: Visualize the Clusters
# Use t-SNE to reduce dimensions for visualization
tsne = TSNE(n_components=2, random_state=42)
latent_2d = tsne.fit_transform(latent_features)
```

```python
# Plot the clusters
plt.figure(figsize=(8, 6))
for i in range(num_clusters):
    plt.scatter(latent_2d[clusters == i, 0], latent_2d[clusters == i, 1], label=f"Cluster {i}")
plt.legend()
plt.title("Clusters of Audio Features")
plt.xlabel("t-SNE Dim 1")
plt.ylabel("t-SNE Dim 2")
plt.show()
```

```python
# Step 7: Save and Evaluate
# Save the encoder and clustering results
np.save("latent_features1.npy", latent_features)
np.save("clusters1.npy", clusters)
# encoder.save("lstm_encoder_model.h5")
encoder.save("lstm_encoder_model1.keras")

print("Clustering complete. Latent features and clusters saved.")
```
