```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text, export_graphviz
from sklearn.metrics import classification_report, accuracy_score
import graphviz
```

```python
df = pd.read_csv('/content/Mental Health Dataset.csv')
```

```python
df.fillna('Unknown', inplace=True)
```

```python
features = ['Gender', 'Occupation', 'family_history', 'Days_Indoors',
            'Growing_Stress', 'Mental_Health_History', 'Mood_Swings',
            'Coping_Struggles', 'Social_Weakness']
target = 'treatment'

# Encode categorical variables
le = LabelEncoder()
df[target] = le.fit_transform(df[target])
for col in features:
    if df[col].dtype == 'object':
        df[col] = le.fit_transform(df[col])
```

```python

# Split data
X = df[features]
y = df[target]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Create and train the decision tree
dt_model = DecisionTreeClassifier(
    max_depth=5,        # Limit tree depth for interpretability
    min_samples_split=20,
    min_samples_leaf=10,
    random_state=42
)
dt_model.fit(X_train, y_train)

```

```python
# Evaluate the model
y_pred = dt_model.predict(X_test)
print("Decision Tree Performance:")
print(classification_report(y_test, y_pred))
print(f"Accuracy: {accuracy_score(y_test, y_pred):.2f}")
```

```python
# Create visual representation (requires graphviz installed)
dot_data = export_graphviz(
    dt_model,
    out_file=None,
    feature_names=X.columns,
    class_names=['No Treatment', 'Treatment'],
    filled=True,
    rounded=True,
    special_characters=True
)

graph = graphviz.Source(dot_data)
graph.render("mental_health_decision_tree")  # Saves as PDF
graph  # Display in notebook if using Jupyter
```

```python
# Get feature importances
importances = dt_model.feature_importances_
feature_importance = pd.DataFrame({
    'Feature': X.columns,
    'Importance': importances
}).sort_values('Importance', ascending=False)

print("\nFeature Importance:")
print(feature_importance)
```
